```python
from sklearn.svm import SVR,SVC
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
import numpy as np
from sklearn.datasets import make_moons, make_circles, make_classification
import matplotlib.pyplot as plt

X, y = make_classification(n_features=2, n_redundant=0, n_informative=2, random_state=1, n_clusters_per_class=1
)


X = (X - X.mean(axis=0, keepdims=True)) / X.std(axis=0, keepdims=True)

print(X.shape)

regr = SVC(C=1e8, kernel='linear')
regr = regr.fit(X, y)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=['r' if _y_ > 0 else 'b' for _y_ in y])
```

```python
regr.dual_coef_
```

```python
regr.coef_.shape, regr.coef_, np.linalg.norm(regr.coef_), regr.coef0
```

```python
xx, yy = np.meshgrid(np.linspace(-4, 4, 100), np.linspace(-4, 4, 100), indexing='ij')

Z = regr.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.pcolor(xx, yy, Z, cmap='coolwarm')
plt.scatter(X[:, 0], X[:, 1], c=['r' if _y_ > 0 else 'b' for _y_ in y])

a = np.linspace(-2, 2, 100).reshape(1, 100)
line = np.array([regr.coef_[0][1], -regr.coef_[0][0]]).reshape(2, 1) @ a # 2, 100

plt.plot(line[0, :], line[1, :], 'r--')

plt.show()
```
